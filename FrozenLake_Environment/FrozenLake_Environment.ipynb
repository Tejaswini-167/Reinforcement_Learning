```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jan 22 09:08:22 2026

@author: Dell
"""
#  import libraries
import gymnasium as gym
import time



# 2.Create and visualize environment
env = gym.make("FrozenLake-v1",render_mode="human")
env.reset()
env.render()


# State space
print("No.of states: ",env.observation_space)

# Action Space
print("No. of actions: ",env.action_space)

# State Transition Probability
print("Transition Probability: ",env.unwrapped.P[0][2])



# deterministic policy and its Visualization
env.step(1)
env.render()



# Random policy and its visualization
a=env.action_space.sample()
env.step(a)
env.render()



# Generate an episode with 20 timespan
env.reset()
ret = 0
num_steps = 20
for i in range(num_steps):
    rnd = env.action_space.sample()
    (next_state, reward, done, trans_prob, info) = env.step(rnd)
    ret = ret + reward
    time.sleep(1)
    env.render()

    if done:
        break
print("Return is",ret)



# Generate 10 such episodes
num_episodes = 5
for i in range(num_episodes):
    env.reset()
    ret = 0
    num_steps = 20
    policy = []
    for i in range(num_steps):
        rnd = env.action_space.sample()
        if rnd == 0:
            policy.append('left')
        elif rnd == 1:
            policy.append('down')
        elif rnd == 2:
            policy.append('right')
        elif rnd == 3:
            policy.append('up')

        (next_state, reward, done, trans_prob, info) = env.step(rnd)

        time.sleep(1)
        env.render()
        ret = ret + reward

        if done:
            break
       # print("Return of episode {0} is {1}".format(i,ret))
        print("Episode# {0}\t Policy: {1}\t Ret: {2}".format(i,policy,ret))

```
